```python
from qiskit import QuantumCircuit
```

```python
from qiskit_aer import AerSimulator

def simulate(qc):
	sim = AerSimulator()
	result = sim.run(qc, shots=500).result()
	counts = result.get_counts()
	print(counts)
```

```python
oracle_1 = QuantumCircuit(2)

oracle_1.h(0)
oracle_1.cx(0,1)


oracle_1.draw()
```

```python
oracle_2 = QuantumCircuit(2)

oracle_2.h(0)
oracle_2.ccx(0,1)

oracle_2.draw()
```

```python
qc = QuantumCircuit(2, 2)

qc.h([0,1])
qc.x(0)
qc.cx(0,1)
qc.measure([0,1], [0,1])
simulate(qc)
print(qc.draw())
```
